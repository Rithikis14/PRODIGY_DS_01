**CLEAR DATA AND VIEW THE TOP 10 MOST POPULAR COUNTRIES IN 2020**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/fe94d34a-0cec-410d-9a7b-964a46c89f4d_Data.csv')

# Extract the relevant columns: 'Country Name' and '2020 [YR2020]'
df_2020 = df[['Country Name', '2020 [YR2020]']]

# Clean the data: Remove rows where population data is missing
df_2020 = df_2020.dropna()

# Rename columns for easier access
df_2020 = df_2020.rename(columns={'Country Name': 'Country', '2020 [YR2020]': 'Population'})

# Sort the data to get the top 10 most populous countries in 2020
df_2020 = df_2020.sort_values(by='Population', ascending=False).head(10)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(df_2020['Country'], df_2020['Population'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population in 2020')
plt.title('Top 10 Most  Countries in 2020')
plt.xticks(rotation=45)
plt.show()
```
